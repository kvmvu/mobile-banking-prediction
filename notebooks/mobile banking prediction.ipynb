# EDA

```python
"""Data types of the various columns"""
train.dtypes
```

```python
"""Plot a histogram grid of the numeric features"""
train.hist(figsize=(14,14), xrot=45)
plt.show()
```

```python
"""Display the numeric features’ summary statistics"""
train.describe()
```

```python
"""Summary statistics of the Categorical features"""
train.describe(include='object')
```

```python
"""Plot categorical distributions"""
for col in train.select_dtypes(exclude='object'):
    if train[col].nunique() <= 10:
        sns.countplot(y=col, data=train)
        plt.show()
```

```python
"""Segment the target by numeric features"""
# num_cols = train.columns[1:4]
# for col in num_cols:
#     sns.boxplot(y = train['Target'].astype('category'), x = col, data=train)
#     plt.show()
```

```python
"""Tabulate target across various categorical features using cross tab"""
for col in train.select_dtypes(exclude='object'):
    if train[col].nunique() <=8:
        display(pd.crosstab(train['Target'], train[col], normalize='index'))
```

```python
"""Count plot of target across various categorical features"""
for col in train.select_dtypes(exclude='object'):
    if train[col].nunique() <= 8:
        g = sns.catplot(x = col, kind='count', col = 'Target', data=train, sharey=False)
        g.set_xticklabels(rotation=60)
```

```python
"""Correlation matrix for the various numeric features"""
corr = train.corr()
plt.figure(figsize=(15,15))
sns.heatmap(corr, cmap='RdBu_r', annot=True, vmax=1, vmin=-1)
plt.show()
```
